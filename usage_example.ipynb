```python
from mycolorpy import colorlist as mcp
import numpy as np
from matplotlib import pyplot as plt
from numpy import random

```

# Creating a list of colors from a given colormap

```python
'''%%!
This example shows how to obtain an equally distributed set of colors from a given colormap.
The function gen_color takes two arguments: cmap and n
cmap is any cmap value from matplotlib: https://matplotlib.org/stable/tutorials/colors/colormaps.html
n= number of color you want from that colormap range
'''
color1=mcp.gen_color(cmap="winter",n=5)
print(color1)
#plt.scatter(color1,color1,c=color1)
color2=mcp.gen_color(cmap="bwr",n=16)
plt.scatter(color2,color2,c=color2)
print(color2)
```

# Creating a list of colors normailized to a given data in an numpy array

```python
a=random.randint(1000, size=(200))
a=np.array(a)
color1=mcp.gen_color_normalized(cmap="seismic",data_arr=a)
plt.scatter(a,a,c=color1)
```

# With reverse

```python
color1=mcp.gen_color_normalized(cmap="seismic",data_arr=a,reverse=True)
plt.scatter(a,a,c=color1)
```
