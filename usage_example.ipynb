```python
import mycolorpy as mcp
import numpy as np
from matplotlib import pyplot as plt
from numpy import random

```

# Example 1

```python
'''%%!
This example shows how to obtain an equally distributed set of colors from a given colormap.
The function gen_color takes two arguments: see_map and n
see_map is any cmap value from matplotlib: https://matplotlib.org/stable/tutorials/colors/colormaps.html
n= number of color you want from that colormap range
'''
color1=mcp.gen_color(see_map="winter",n=5)
print(color1)
#plt.scatter(color1,color1,c=color1)
color2=mcp.gen_color(see_map="bwr",n=16)
plt.scatter(color2,color2,c=color2)
print(color2)
```
